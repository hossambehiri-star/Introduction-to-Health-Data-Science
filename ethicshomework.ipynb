### Find another example of the misuse of health data. Give details about the event and discuss the ethical issues involved. What could have been done to prevent this from happening?

One example of the misuse of health data is the 2019 case of Google partnering with Ascension Health in what was known as Project Nightingale. In this project, Google collected and analyzed the medical records of millions of patients, including sensitive information such as names, diagnoses, and lab results, without informing the patients. This raised major ethical concerns, particularly around the lack of informed consent and transparency, as patients were not aware their data was being shared. It also raised issues of privacy and confidentiality, since sensitive health information was given to a large technology company, with concerns it could be used for purposes beyond patient care. To prevent this, healthcare organizations should ensure patients are fully informed and give consent before their data is shared, use anonymized data whenever possible, and implement stronger data protection policies and regulations to safeguard patient privacy.


### Find another example of a breach in health data. Describe the event, including the number of people affected and the consequences. What could have been done to prevent this from happening?

One example of a health data breach is the Anthem Inc. data breach in 2015, one of the largest in U.S. history. Hackers gained unauthorized access to the health insurer’s database, exposing the personal information of about 78.8 million people, including names, birth dates, Social Security numbers, and medical IDs. Although medical records were not directly accessed, the stolen data could be used for identity theft and fraud. The breach resulted in major financial losses, lawsuits, and a $16 million settlement for violating HIPAA regulations. This could have been prevented with stronger cybersecurity measures, such as encrypting sensitive data, improving system monitoring, and using multi-factor authentication to limit unauthorized access.

### Find another example of bias in health data. Describe the event and the consequences. What could have been done to prevent this from happening?
One example of bias in health data is a widely used U.S. healthcare algorithm that was found to underestimate the health needs of Black patients. The algorithm used healthcare costs as a proxy for illness, assuming that patients who spend more money are sicker. However, due to systemic inequalities, Black patients often have less access to care and therefore lower healthcare spending, even when they are equally or more ill. As a result, the algorithm incorrectly rated Black patients as lower risk, leading to fewer referrals for extra care and resources compared to white patients. This reinforced existing health disparities and limited access to needed treatment. To prevent this, developers should use more accurate measures of health (such as clinical data instead of cost), regularly audit algorithms for bias, and include diverse populations when designing and testing healthcare systems.

### Find an example of anonymization of health data. Describe the technique used.
One example of anonymization of health data is when hospitals or researchers remove personally identifiable information (PII) from patient records before sharing them for research. This process is called de-identification. It involves removing direct identifiers such as names, addresses, Social Security numbers, and medical record numbers, and sometimes generalizing data (for example, using age ranges instead of exact ages or removing exact dates). In the United States, this is often done following HIPAA Safe Harbor standards, which require removing 18 types of identifiers. By doing this, the data can be used for research and analysis without being linked back to specific individuals, helping protect patient privacy.

Studies showed pulse oximeters can be less accurate in people with darker skin, leading to underdiagnosis of low oxygen levels.

### What is the role of an Institutional Review Board (IRB) in research?
An IRB reviews research involving humans to ensure it follows ethical standards, minimizes risk, ensures informed consent, and protects participant rights and confidentiality.


### What is the difference between spreadsheet software and database software for health data management?

Spreadsheets (like Excel) are useful for small, simple datasets, while databases are designed for large, complex datasets, offering better data security, organization, and the ability to run advanced queries.

### What is a conflict of interest in public health research?
A COI occurs when a researcher’s financial, personal, or professional interests could influence or bias their research decisions or outcomes.

### What is informed consent in the context of health data collection?
Participants must be fully informed about the purpose of the research, risks, benefits, and how their data will be used, and must voluntarily agree to participate.


### What is data masking in the context of de-identification of health data?
A de-identification method where sensitive information is replaced or hidden (e.g., replacing names with codes) to reduce the risk of identification.

### What is the principle of  k -anonymity in the context of de-identification of health data?
A technique that ensures each individual in a dataset is indistinguishable from at least k other individuals, reducing the chance of re-identification.


### What is the main concern about conflicts of interest in public health research?

Conflicts of interest can bias research findings, leading to inaccurate results and loss of trust in public health recommendations.

### What is the first step in managing conflicts of interest in public health research?
The first step is full disclosure of any potential conflicts to institutions, reviewers, and the public.

### What is the role of independent oversight in managing conflicts of interest in public health research?
Independent oversight (such as ethics boards or external reviewers) helps ensure research is objective, transparent, and not influenced by personal interests.

### What is one strategy for mitigating conflicts of interest in public health research?

Strategies include separating financial interests from research decisions, using independent data analysis, and requiring transparency.


### What is the potential impact of not properly managing conflicts of interest in public health research?
Failure to manage conflicts can lead to biased or misleading research, harm to patients, poor policy decisions, and loss of public trust in science and healthcare.
